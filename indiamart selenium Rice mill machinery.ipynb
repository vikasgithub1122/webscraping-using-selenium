```python
def find_links():
    #Find all links on the page
    links = driver.find_elements(By.XPATH, '//a[@class="fs18 ptitle "]')
     # h = [link.get_attribute("href") for link in links]
    links = [i.get_attribute('href') for i in links]
    return links
```

```python
print(len(find_links()))
print("Links found:", find_links())
```

```python
def find_names():
    #Find all names
    names = driver.find_elements(By.XPATH, '//h2[@class="lcname"]')   
    name = [i.text for i in names]
        #find_name().append(i.text)
    return name   
```

```python
print(len(find_names()))
print("names found:", find_names())
```

```python
 def find_locations():
    # Find locations, assuming they're in spans with a specific class
    locations = driver.find_elements(By.XPATH, '//p[@class="sm clg"]')
    return [location.text for location in locations]
```

```python
print(len(find_locations()))
print("locations found:", find_locations())
```

```python
def find_phone_numbers():
    phone_elements = driver.find_elements(By.XPATH, '//span[@class="pns_h duet"]')
    return [element.get_attribute("innerHTML") for element in phone_elements]
```

```python
print(len(find_phone_numbers()))
print("phone numbers found:", find_phone_numbers())
```

```python
def find_product_names():
    product_names = driver.find_elements(By.TAG_NAME, 'h3')
    product = []
    for i in product_names[1:]:
        product.append(i.text)
    return product
```

```python
print(len(find_product_names()))
print("product names found:", find_product_names())
```
